## 1. Objective

Module A compares a custom B+ Tree indexing structure against a brute-force baseline. The objective is to demonstrate that the B+ Tree supports efficient insert, search, delete, update, and range-query operations while remaining suitable for lightweight DBMS-style usage.


## 2. Implementation Summary

Core files:

- `database/bplustree.py`: custom B+ Tree implementation
- `database/bruteforce.py`: linear-scan baseline
- `database/table.py`: table wrapper for record operations
- `database/db_manager.py`: manager for multiple tables
- `benchmark.py`: automated timing and memory benchmark

The B+ Tree uses ordered internal nodes for routing and leaf-level chaining for range queries. This allows point lookups in logarithmic time and efficient range scans using linked leaves.


## 3. Theoretical Complexity

- Insert: approximately `O(log n)`
- Search: approximately `O(log n)`
- Delete: approximately `O(log n)`
- Range query: approximately `O(log n + k)` where `k` is the number of returned records
- Brute-force search baseline: `O(n)`

This is why the B+ Tree is expected to scale better than a linear structure as the number of records increases.

## 6. Observations

- Search performance improved because the B+ Tree navigates through ordered keys and reaches the target leaf in logarithmic time, while the brute-force baseline may inspect a large fraction of stored records.
- Range query performance improved because the B+ Tree can locate the starting key quickly and then continue through linked leaf nodes, whereas the brute-force approach must scan the full dataset and then sort matching values.
- Insert and delete operations remained efficient because rebalancing is localized to affected nodes. This introduces some structural maintenance overhead, but it scales better than repeated full-list operations as record count grows.
- Memory usage is a tradeoff: the B+ Tree stores additional structural metadata such as internal nodes and leaf links, so it may consume more bookkeeping space than a simple list. In return, it provides faster lookup and range access for larger datasets.
- Overall, the benchmark plots support the expected result that the B+ Tree becomes more advantageous as dataset size increases, especially for search-heavy and range-query workloads.


## 7. Conclusion

The custom B+ Tree provides a better indexing strategy than brute-force linear scanning for growing datasets, especially for search and range-query workloads. This makes it suitable as the indexing core of a lightweight DBMS prototype.
